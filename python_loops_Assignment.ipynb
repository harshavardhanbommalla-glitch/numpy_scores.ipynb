```python
import numpy as np

# 1. Create the NumPy array
temps_celsius = np.array([22, 25, 28, 24, 26])

# 2. Convert to Fahrenheit (F = C * 1.8 + 32)
temps_fahrenheit = (temps_celsius * 1.8) + 32

# 3. Print both arrays
print(f"Celsius:    {temps_celsius}")
print(f"Fahrenheit: {temps_fahrenheit}")

# 4. Calculate average in Fahrenheit
avg_f = np.mean(temps_fahrenheit)
print(f"Average Temperature: {avg_f}°F")

scores = np.array([85, 90, 78, 92, 88, 76, 95, 82, 89, 91, 87, 84])

# 1. Print shape and total elements
print(f"Array Shape: {scores.shape}")
print(f"Total Elements: {scores.size}")

# 2. Calculate Stats
highest = scores.max()
lowest = scores.min()
score_range = highest - lowest

print(f"Highest Score: {highest}")
print(f"Lowest Score:  {lowest}")
print(f"Score Range:   {score_range}")

import time

# 1. Create Data
num_elements = 50000
np_arr = np.arange(1, num_elements + 1)
py_list = list(range(1, num_elements + 1))

# 2. Time NumPy
start_np = time.time()
np_sum = np.sum(np_arr)
end_np = time.time()
np_time = end_np - start_np

# 3. Time Python List
start_py = time.time()
py_sum = sum(py_list)
end_py = time.time()
py_time = end_py - start_py

# 4. Results
print(f"NumPy Sum: {np_sum} | Time: {np_time:.6f}s")
print(f"Python Sum: {py_sum} | Time: {py_time:.6f}s")

speed_factor = py_time / np_time
print(f"NumPy is {speed_factor:.2f}x faster than Python lists for this operation.")
```
